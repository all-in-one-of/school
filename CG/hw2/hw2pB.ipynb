```python
%matplotlib inline
import skdemo
import matplotlib.pyplot as plt
import numpy as np
from skimage import io
from scipy.ndimage import convolve
import skimage.filter as filters

leaf = io.imread('leaf.jpg')
r = io.imread('r.jpg')
```

```python
#Create a kernel that blurs and darken an image
#I used mean_kernel similar to the tutorial
#it seems pretty effective in terms returning the desired result
#kernel is used in creating new filters
#all other methods are used in creating new filters as well (gaussian, pixelated...)
def kernel(img):
    image = skimage.color.rgb2gray(img)
    step_signal = np.zeros(100)
    step_signal[25:] = 10
    step_signal[75:] = 0
    plt.plot(step_signal)
    plt.margins(0.1)

    noisy_signal = np.copy(step_signal)
    noisy_signal[::2] += 1
    plt.plot(noisy_signal)
    plt.margins(0.1)

    # Take the mean of neighboring pixels
    smooth_signal = (noisy_signal[:-1] + noisy_signal[1:]) / 2.0
    plt.plot(smooth_signal)
    plt.margins(0.1)

    # Same as above, using a convolution kernel
    # Neighboring pixels multiplied by 1/2 and summed
    mean_kernel = np.array([1, 1]) / 2.0
    smooth_signal = np.convolve(noisy_signal, mean_kernel, mode='valid')
    plt.plot(smooth_signal)
    plt.margins(0.1)

    mean_kernel = 1.0/10.0 * np.ones((3, 3))

    print mean_kernel

    filtered = convolve(image, mean_kernel)
    skdemo.imshow_all(image, filtered)
    
kernel(r)
```

```python
#DOWNSAMPLE
#pixelate an image with given pixels
#used in pixelating an image 

def ds(image, px):
    pixelated = image[::px, ::px]
    skdemo.imshow_all(image, pixelated)
    
ds(leaf, 50)
ds(r, 10)
```

```python
#GAUSSIAN
#Blur the image using Gaussian operations
#this method did a good job bluring images
def gaus(image):
    size = 5
    structuring_element = np.ones((3*size, 3*size, 3))
    smooth_gaussian = filters.gaussian_filter(image, size, multichannel=True)
    skdemo.imshow_all(image, smooth_gaussian)
gaus(leaf)
gaus(r)
```

```python
#HORIZONTAL AND VERTICAL EDGE DETECTIONS
#identify horizontal and vertical edges of an image

def ed(image, px):
    img = skimage.color.rgb2grey(image)
    horizontal_edge_kernel = np.array([[10], [0], [-10]])
    vertical_edge_kernel = np.array([[10, 0, -10]])
    pixelated = img[::px, ::px]
    pixelated = skimage.img_as_float(pixelated)
    vertical_gradient = convolve(pixelated, horizontal_edge_kernel)
    horizontal_gradient = convolve(pixelated, vertical_edge_kernel)
    skdemo.imshow_all(horizontal_gradient, vertical_gradient)
    
ed(leaf, 5)
ed(r, 2)
    
```
