```python
from skimage import data
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
from skimage import io
leaf = io.imread('leaf.jpg')
sun = io.imread('sun.jpg')
```

```python
# HISTOGRAM : tonal distribution of an image
# used in analyzing image tones and adjusting them

def htg(img):
    skdemo.imshow_with_histogram(img);
    
htg(leaf)
htg(sun)
```

```python
# THRESHOLD: can be used in edge detection, image segmentation
#hard to identify a threshold in a colorful image
from skimage.filter import threshold_otsu
def thr(image):
    threshold = threshold_otsu(image)
    ax_image, ax_hist = skdemo.imshow_with_histogram(image)
    ax_image.imshow(image > threshold)
    ax_hist.axvline(threshold, color='red');

thr(leaf)
thr(sun)
```

```python
#H, V, S: Hue , Saturation and Value
#one of the most popular RGB color representation
from skimage import color
def translate_to_hsv(image):
    #rgb_array = np.shape(image)
    hsv = color.rgb2hsv(image)
    print hsv;
print("---Leaf---")
translate_to_hsv(leaf)
print("---Sun---")
translate_to_hsv(sun)
```

```python
#CIELAB:color-opponent space. L for lightness, a and b are color-opponent dimension
#Usede for isolating objects, green screen
def cielab(image):
    lab = color.rgb2lab(image)
    #L = lab[:, :, 0]
    #a = lab[:, :, (0, 2)] = 0
    #b = lab[:, :, (0, 1)])
    #plt.imshow(L);
    print lab.shape;
    skdemo.imshow_all(*np.rollaxis(lab, axis=-1));
cielab(sun)
cielab(leaf)
    
```
