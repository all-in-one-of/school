```python
import numpy as np
from matplotlib import pyplot as plt, cm
%matplotlib inline
import skdemo
from skimage import data, color
from skimage import io
from skimage import morphology

leaf = io.imread('leaf.jpg')
r = io.imread('bubbles.jpg')

#Remove noise using morphological operations
#morphological operation - a technique used in analyzing shapes
#2 main morphological operations: dilation, erosion
#here in order to remove noise, I had to first find the main object,
#which is a bubble with a radius of 10,
#do an opening around that bubble so everything outside that 
#bubble got removed
#This method is efficient if the main object is actually standing out in the picture,
#otherwise, it's hard to detect which one is your main bubble
def remnoise(image):
    hub = color.rgb2gray(image [90:200, 50:150], )
    plt.imshow(hub);
    
    disk = morphology.disk(radius=10)
    gal = morphology.opening(hub, disk)
    skdemo.imshow_with_histogram(gal)
    
remnoise(r)
```
