```python
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
import skdemo
from skimage import io, segmentation as seg, color
from skimage import data, filter as filters
from matplotlib import pyplot as plt, cm
from scipy import ndimage as nd
from skimage import feature
from skimage.morphology import watershed
from skimage.measure import regionprops

#plt.rcParams['image.cmap'] = 'spectral'

leaf = io.imread('leaf.jpg')
r = io.imread('r.jpg')
b = io.imread('bubbles.jpg')
```

```python
#SLIC
#Used in segmentation based on color and space
#This algorithm did a petty good job identifying each area main color and create k segments of them
#SLIC can be used in edge detection/object recognition
def slic(image):
    labels = seg.slic(image, n_segments=25, compactness=20, 
                      sigma=2, enforce_connectivity=True)
    out = np.zeros_like(image)
    for label in np.unique(labels):
        indices = np.nonzero(labels == label)
        out[indices] = np.mean(image[indices], axis=0)
    return out

skdemo.imshow_all(leaf, slic(leaf))
skdemo.imshow_all(r, slic(r))
```

```python
#watershed is another segmentation algorithm, implemented by detecting edges and then segment them
#It is not as efficient as SLIC because edges sometimes are hard to define
#just like SLIC, watershed can be used in object detection
def wtr(image):
    y = np.array([1, 0, 1, 2, 1, 3, 2, 0, 2, 4, 1, 0])

    seeds = nd.label(y == 0)[0]
    
    img = color.rgb2gray(image)
    edges = filters.sobel(img)
    plt.imshow(edges, cmap='gray');
    
    
    threshold = 0.5

    # Euclidean distance transform
    # How far do we ave to travel from a non-edge to find an edge?
    non_edges = (edges < threshold)
    distance_from_edge = nd.distance_transform_edt(non_edges)
    
    peaks = feature.peak_local_max(distance_from_edge)
    peaks_image = np.zeros(img.shape, np.bool)
    peaks_image[tuple(np.transpose(peaks))] = True
    seeds, num_seeds = nd.label(peaks_image)

    
    ws = watershed(edges, seeds)
    
    regions = regionprops(ws)

    ws_updated = ws.copy()
    for region in regions:
        if region.eccentricity > 0.7:
            ws_updated[ws_updated == region.label] = 0

    plt.imshow(color.label2rgb(ws_updated, image, bg_label=0));

    
wtr(leaf)
wtr(b)
```
