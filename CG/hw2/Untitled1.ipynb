```python
%matplotlib inline
import skdemo
import matplotlib.pyplot as plt
import numpy as np
from skimage import io, color
import skimage
leaf = io.imread('leaf.jpg')
sun = io.imread('sun.jpg')
```

```python
# Create some arbitrary kernel
# Create some arbitrary kernel
def generate_kernel(size):
    kernel = np.zeros((size,size))
    for x in range(size):
        kernel[x][x] = 1.0
        kernel[size-1-x][x] = 1.0
        kernel = kernel / float(sum(sum(kernel)))
    return kernel

kernel = generate_kernel(50)
plt.imshow(kernel)

imgbw = color.rgb2grey(leaf)
plt.figure()
plt.subplot(1, 2, 1)
plt.imshow(imgbw)
plt.subplot(1, 2, 2)
plt.imshow(np.convolve(imgbw, kernel))
```
