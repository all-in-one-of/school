```python
from PIL import Image
from skimage import data
from skimage import color
from skimage import exposure
from skimage import img_as_float
import skimage.filter as filters
from scipy import ndimage
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import skdemo
from skimage.draw import circle, circle_perimeter

from skimage import io
#Original image
original = img_as_float(io.imread('original.jpg'))
#With an instagram filter"toaster"
toaster = img_as_float(io.imread('toaster.jpg'))
skdemo.imshow_all(original, toaster)
```

```python
# add a reddish color to an image
def colorize(image):
    img = image.copy()
    rgb = img
    blue = [0.88, 0.6, 0.77]
    #blue = [0.52, 0.4, 1]
    r = img * blue
    return r
skdemo.imshow_all(original, toaster, colorize(original))

# create a circled vignette filter mask
def filt(image):
    w, h = image.shape[:2]
    
    img = np.ones((w, h, 3), dtype=np.double)
    img = colorize(img)

    # fill circle
    rr, cc = circle(w*0.001, h*0.08, h-20, img.shape)
    img[rr, cc, :] = (1, 1, 1)
    r = filters.gaussian_filter(img, sigma = 100, multichannel=True)

    return r
plt.imshow(filt(original))
    
```

```python
#combine everything into the final filter
def brighter(image):
    color = image.copy()
    f = [1, 0.87, 0.75]
    frame = filt(image)
    r = color * f * frame 
    new = exposure.adjust_gamma(r, gamma=0.56, gain=1)
    new1 = exposure.rescale_intensity(new, in_range=(0.2, 0.9))
    return new1
skdemo.imshow_all(original,toaster, brighter(original))

```
