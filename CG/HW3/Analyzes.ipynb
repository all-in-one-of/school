```python
from PIL import Image
from skimage import data
from skimage import color
from skimage import morphology
from skimage import img_as_float
import skimage.filter as filters
from scipy import ndimage
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import skdemo
from skimage.draw import circle, circle_perimeter
import math

from skimage import io
#Original image
original = img_as_float(io.imread('original.jpg'))
#With an instagram filter"toaster"
toaster = img_as_float(io.imread('toaster.jpg'))
skdemo.imshow_all(original, toaster)
```

```python
#Difference between the original image and the filtered one
def dif(org, fil):
    new = color.rgb2gray(org) - color.rgb2gray(fil)
    return new
plt.imshow(dif(original, toaster))
```

```python
#Comparing edges of the original and filtered images
def canny_edge(org, fil):
    org_canny = filters.canny(color.rgb2gray(org))
    fil_canny = filters.canny(color.rgb2grey(fil))
    o = org_canny[100:400, 0:300]
    f = fil_canny[100:400, 0:300]
    skdemo.imshow_all(o, f)
canny_edge(original, toaster)
```

```python
#HISTOGRAMS
# increase red's intensity
# decrease blue's instensity

skdemo.imshow_with_histogram(original)
skdemo.imshow_with_histogram(toaster)
```

```python
#Using the original image threshold
from skimage.filter import threshold_otsu
def thr(image):
    thresh = threshold_otsu(color.rgb2gray(original))
    img = color.rgb2gray(image) > thresh
    print thresh
    return img
skdemo.imshow_all(thr(original), thr(toaster))
```
